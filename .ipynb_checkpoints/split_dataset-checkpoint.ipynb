```python
import os
import random
import shutil
from sklearn.model_selection import train_test_split

# Set the seed for reproducibility
random.seed(42)

# Paths
annotations_path = r'C:\Users\ucsfg\Documents\Code\boutRight_v3\annotations'
images_path = r'C:\Users\ucsfg\Documents\Code\boutRight_v3\wavfiles\images_py'
dataset_path = r'C:\Users\ucsfg\Documents\Code\boutRight_v3\dataset'
images_dir = os.path.join(dataset_path, 'images')
labels_dir = os.path.join(dataset_path, 'labels')
train_images_dir = os.path.join(images_dir, 'train')
val_images_dir = os.path.join(images_dir, 'val')
train_labels_dir = os.path.join(labels_dir, 'train')
val_labels_dir = os.path.join(labels_dir, 'val')

# Create directories if they don't exist
os.makedirs(train_images_dir, exist_ok=True)
os.makedirs(val_images_dir, exist_ok=True)
os.makedirs(train_labels_dir, exist_ok=True)
os.makedirs(val_labels_dir, exist_ok=True)
```

```python

# Get list of labeled images
labels = [f for f in os.listdir(annotations_path) if f.endswith('.xml')]
images = [f.replace('.xml', '.png') for f in labels]

labels
```

```python
# Split data into train and val sets
train_images, val_images, train_labels, val_labels = train_test_split(images, labels, test_size=0.2, random_state=42)

# Function to copy files
def copy_files(file_list, src_dir, dst_dir):
    for file in file_list:
        src_file = os.path.join(src_dir, file)
        dst_file = os.path.join(dst_dir, file)
        shutil.copy(src_file, dst_file)

# Copy train files
copy_files(train_images, images_path, os.path.join(dataset_path, 'images/train'))
copy_files(train_labels, annotations_path, os.path.join(dataset_path, 'labels/train'))

# Copy val files
copy_files(val_images, images_path, os.path.join(dataset_path, 'images/val'))
copy_files(val_labels, annotations_path, os.path.join(dataset_path, 'labels/val'))

print("Files copied successfully!")
```
